```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8), tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('1. Line Plot (axes[0, 0])')

x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1, color='red')
axes[0, 1].set_title('2. Sin Function (axes[0, 1])')

data2 = np.random.randn(1000)
axes[1, 0].hist(data2, bins=30, alpha=0.7, color='green')
axes[1, 0].set_title('3. Histogram (axes[1, 0])')


x3 = np.random.rand(50)
y3 = np.random.rand(50)
axes[1, 1].scatter(x3, y3, s=y3*200, c=x3, cmap='plasma', alpha=0.6)
axes[1, 1].set_title('4. Scatter Plot (axes[1, 1])')

axes[0, 2].set_title('5. Empty Plot (axes[0, 2])')
axes[1, 2].set_title('6. Empty Plot (axes[1, 2])')

plt.show()
```
