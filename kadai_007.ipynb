```python
import pandas as pd
import matplotlib.pyplot as plt
# japanize_matplotlibはStep1で既にインポート済みと仮定します

# 【重要】dfを定義するコード（ファイルの読み込み）をここで実行します
# ----------------------------------------------------
df = pd.read_csv('category.csv')
# ----------------------------------------------------

# Step 2: カテゴリーの出現頻度をカウントし棒グラフ化
print("--- 1. カテゴリー別出現頻度の棒グラフ ---")
df['カテゴリー'].value_counts().plot(
    kind='bar',
    figsize=(10, 6),
    title='商品カテゴリー別 出現頻度'
)

# Step 3: グラフの表示
plt.show()


# Step 4: グループごとの基本統計量の算出
# エラーが出た箇所の修正と実行
print("\n--- 2. 商品番号ごとのカテゴリー文字数の基本統計量 ---")

# 1. 数値データがないため、「カテゴリー」の文字列長をサブ列として使用します
df['カテゴリー文字数'] = df['カテゴリー'].str.len()

# 2. 「商品番号」を基準列とし、「カテゴリー文字数」のdescribe()を算出します
統計量 = df.groupby('商品番号')['カテゴリー文字数'].describe()

# 結果の表示
print(統計量)
```
